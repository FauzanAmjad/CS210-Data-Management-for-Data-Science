```python
import csv
```

```python
#num 1
def pokemon1():
    reader = csv.DictReader(open('pokemonTrain.csv'))
    #print(reader.fieldnames)
    #print(','.join(reader.fieldnames))
    total = 0
    over = 0
    for row in reader:
        values = list(row.values())  # need to cast row.values() to list
        if 'fire' in values[4]:
            #if values[4] == 'fire':
            total = total + 1
            if float(values[2]) >= 40:
                over = over + 1
    #print(total)
    #print(over)
    avg = float(over/total) * 100
    #print(round(avg))
    with open('pokemon1.txt', 'w') as f:
        f.write('Percentage of fire type pokemon at or above level 40 = ' + str(round(avg)))
    f.close()
pokemon1()
```

```python
#num2
def pokemon2_and_3():
    reader = csv.DictReader(open('pokemonTrain.csv'))
    weakness_dict = {}
    for row in reader :
        r = list(row.values())
        if 'NaN' in r[4] :
            continue
        if(r[5] in weakness_dict.keys()):
            weakness_dict[r[5]].append(r[4])
        else :
            weakness_dict[r[5]] = [r[4]]
    #print(weakness_dict)
    for k,v in weakness_dict.items() :
        freq = {}
        for _type in v :
            if _type in freq :
                freq[_type] = freq[_type] + 1
            else :
                freq[_type] = 1
        #print(freq)
        max_val = max(list(freq.values()))
        #print(max_val)
        indices = [index for index, val in enumerate(list(freq.values())) if val == max_val]
        #print(indices)
        if len(indices) == 1 :
            weakness_dict.update({k:v[indices[0]]})
        else :
            list_to_sort = []
            for index in indices :
                list_to_sort.append(v[index])
                #print(list_to_sort)
            list_to_sort.sort()
            weakness_dict.update({k:list_to_sort[0]})
    #print(weakness_dict)

    #num3
    reader = csv.DictReader(open('pokemonTrain.csv'))
    #indices atk 6, def 7, hp 8
    #atk variables
    atk41total = 0
    atk39total = 0
    atk41count = 0
    atk39count = 0

    #def variables
    def41total = 0
    def39total = 0
    def41count = 0
    def39count = 0

    #hp variables
    hp41total = 0
    hp39total = 0
    hp41count = 0
    hp39count = 0

    for row in reader :
        r = list(row.values())
        if not 'NaN' in r[6] :
            if(float(r[2]) > 40):
                atk41total = atk41total + float(r[6])
                atk41count = atk41count + 1
            elif(float(r[2]) <= 40):
                atk39total = atk39total + float(r[6])
                atk39count = atk39count + 1
        if not 'NaN' in r[7] :
            if(float(r[2]) > 40):
                def41total = def41total + float(r[7])
                def41count = def41count + 1
            elif(float(r[2]) <= 40):
                def39total = def39total + float(r[7])
                def39count = def39count + 1
        if not 'NaN' in r[8] :
            if(float(r[2]) > 40):
                hp41total = hp41total + float(r[8])
                hp41count = hp41count + 1
            elif(float(r[2]) <= 40):
                hp39total = hp39total + float(r[8])
                hp39count = hp39count + 1

    atk41avg = float(atk41total/atk41count)
    atk41avg = round(atk41avg, 1)
    atk39avg = float(atk39total/atk39count)
    atk39avg = round(atk39avg, 1)
    def41avg = float(def41total/def41count)
    def41avg = round(def41avg, 1)
    def39avg = float(def39total/def39count)
    def39avg = round(def39avg, 1)
    hp41avg = float(hp41total/hp41count)
    hp41avg = round(hp41avg, 1)
    hp39avg = float(hp39total/hp39count)
    hp39avg = round(hp39avg, 1)
    #print("atk lvl > 40")
    #print(atk41avg)
    #print(atk41total)
    #print(atk41count)
    #print("atk lvl <= 40")
    #print(atk39avg)
    #print(atk39total)
    #print(atk39count)
    #print("def lvl > 40")
    #print(def41avg)
    #print(def41total)
    #print(def41count)
    #print("def lvl <= 40")
    #print(def39avg)
    #print(def39total)
    #print(def39count)
    #print("hp lvl > 40")
    #print(hp41avg)
    #print(hp41total)
    #print(hp41count)
    #print("hp lvl <= 40")
    #print(hp39avg)
    #print(hp39total)
    #print(hp39count)
    #nums 2 and 3 writing
    reader = csv.DictReader(open('pokemonTrain.csv'))
    resultFile = open('pokemonResult.csv', 'w', newline = '')
    writer = csv.writer(resultFile)
    writer.writerow(reader.fieldnames)
    for row in reader :
        values = list(row.values())
        #print(values)
        if values[4] == 'NaN' :
            values[4] = weakness_dict.get(values[5])
        if values[6] == 'NaN' :
            if float(values[2]) > 40:
                values[6] = atk41avg
            elif float(values[2]) <= 40:
                values[6] = atk39avg
        if values[7] == 'NaN' :
            if float(values[2]) > 40:
                values[7] = def41avg
            elif float(values[2]) <= 40:
                values[7] = def39avg
        if values[8] == 'NaN' :
            if float(values[2]) > 40:
                values[8] = hp41avg
            elif float(values[2]) <= 40:
                values[8] = hp39avg
        writer.writerow(values)
    resultFile.close()
pokemon2_and_3()
```

```python
#num 4
def pokemon4():
    reader = csv.DictReader(open('pokemonResult.csv'))
    type_nature_dict = {}
    for row in reader :
        r = list(row.values())
        if not (r[4] in type_nature_dict.keys()) :
            type_nature_dict[r[4]] = [r[3]]
        else :
            if r[3] in type_nature_dict.get(r[4]) :
                continue
            else :
                type_nature_dict[r[4]].append(r[3])

    sorted_type_nature_dict = dict(sorted(type_nature_dict.items()))

    result = 'Pokemon type to personality mapping:\n\n'
    for k,v in sorted_type_nature_dict.items() :
        result += '   ' + k + ': '
        #print(k)
        v.sort()
        for nature in v :
            #print(v)
            if nature == v[-1] :
                result += nature + '\n'
            else :
                result += nature + ', '

    with open('pokemon4.txt', 'w') as f:
        f.write(result)
    f.close()
pokemon4()
```

```python
#num 5
def pokemon5():
    reader = csv.DictReader(open('pokemonResult.csv'))
    total = 0
    over = 0
    for row in reader:
        values = list(row.values()) 
        if float(values[9]) == 3:
            #print(values[1])
            total = total + float(values[8])
            over = over + 1
    #print(total)
    #print(over)
    avg = (total/over)
    with open('pokemon5.txt', 'w') as f:
        f.write('Average hit point for pokemon of stage 3.0 = ' + str(round(avg)))
    f.close()
pokemon5()
```
