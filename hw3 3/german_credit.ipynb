```python
fileName = 'GermanCredit.csv'
```

Preprocesing

P.1

```python
def dropCols(n = 3):
    data = pd.read_csv(fileName)
    count = data.eq('none').sum().to_frame()
    count = count.nlargest(n, 0)
    data = data.drop(count.index.values, axis = 1)
    return data
data = dropCols()
data.head(60)
```

P.2

```python
def removeApos():
    data = pd.read_csv(fileName)
    for row in data.index:
        for col in range(len(data.columns)):
            item = str(data.iloc[row, col])
            data.iloc[row, col] = item.replace("'", "")
    return data
data = removeApos()
data.head(60)
```

P.3

```python
def renameCheckStat(d=None):
    if (d is None):
        data = pd.read_csv(fileName)
    else:
        data = d
    newCol = []
    renameDict = {'no checking': 'No Checking', '<0': 'Low', '0<=X<200': 'Medium', '>=200': 'High'}
    for i in data['checking_status']:
        newCol.append(renameDict[i.replace("'", "")])
    data['checking_status'] = newCol
    return data
data = renameCheckStat()
data.head(60)
```

P.4

```python
def renameSaveStat(d=None):
    if (d is None):
        data = pd.read_csv(fileName)
    else:
        data = d
    newCol = []
    renameDict = {'no known savings': 'No Savings', '<100': 'Low', '100<=X<500': 'Medium', '500<=X<1000': 'High', '>=1000': 'High'}
    for i in data['savings_status']:
        newCol.append(renameDict[i.replace("'", "")])
    data['savings_status'] = newCol
    return data
data = renameSaveStat()
data.head(60)
```

P.5

```python
def renameCE(d=None):
    if (d is None):
        data = pd.read_csv(fileName)
    else:
        data = d
    newCCol = []
    newECol = []
    renameCDict = {'good': '1', 'bad': '0'}
    renameEDict = {'unemployed': 'Unemployed', '<1': 'Amateur', '1<=X<4': 'Professional', '4<=X<7': 'Experienced', '>=7': 'Expert'}
    for i in data['employment']:
        newECol.append(renameEDict[i.replace("'", "")])
    for i in data['class']:
        newCCol.append(renameCDict[i.replace("'", "")])
    data['employment'] = newECol
    data['class'] = newCCol
    return data
data = renameCE()
data.head(60)
```

A.2

```python
def avgCredit():
    data = removeApos()
    data = renameCE(data)
    credits = data[(data['employment'] == 'Experienced') & (data['personal_status'] == 'male single')]['credit_amount']
    return np.asarray(credits, dtype=np.float64).mean()
avgCredit = avgCredit()
print('average credit amount of single males with 4<=X<7 experience: ' + str(avgCredit))
```

A.3

```python
def avgJobCreditTime():
    print("Credit Durations By Job Type")
    data = removeApos()
    skilledAvgDur = np.asarray(data[data['job'] == 'skilled']['duration'], dtype=np.float64).mean()
    print('skilled: ' + str(skilledAvgDur))
    unskilledAvgDur = np.asarray(data[data['job'] == 'unskilled resident']['duration'], dtype=np.float64).mean()
    print('unskilled resident: ' + str(unskilledAvgDur))
    highAvgDur = np.asarray(data[data['job'] == 'high qualif/self emp/mgmt']['duration'], dtype=np.float64).mean()
    print('high qualif/self emp/mgmt: ' + str(highAvgDur))
    unESNonResAvgDur = np.asarray(data[data['job'] == 'unemp/unskilled non res']['duration'], dtype=np.float64).mean()
    print('unemp/unskilled non res: ' + str(unESNonResAvgDur))
avgJobCreditTime()
```

```python
def edStats():
    data = removeApos()
    mostCheckStat = data[data['purpose'] == 'education']['checking_status'].mode().iloc[0]
    print('Most common checking status: ' + str(mostCheckStat))
    mostCheckStat = data[data['purpose'] == 'education']['savings_status'].mode().iloc[0]
    print('Most common savings status: ' + str(mostCheckStat))
edStats()
```

```python
data = pd.read_csv(fileName)
greater4k = data[data['credit_amount'].astype(int) > 4000][['property_magnitude', 'age']]
grouped = greater4k.groupby('property_magnitude').mean()
grouped = grouped.rename(index = lambda i: i.replace("'", ""))
grouped = grouped.reset_index()
plt.xlabel('Property Magnitude')
plt.ylabel('Average Age')
plt.bar(grouped['property_magnitude'], grouped['age'], width = 0.3)
plt.show()
```

```python
data = removeApos()
data = renameSaveStat(data)
fig, ref = plt.subplots(3, 1, figsize = (10, 30))
data = data[(data['age'].astype(int) > 40) & (data['savings_status'] == 'High')]

pers = data.groupby('personal_status')['savings_status'].count()
pers = pers.reset_index()
ref[0].pie(pers['savings_status'], labels = pers['personal_status'])
ref[0].set_title("a. High Savings Breakdown by Personal Status")

cred = data.groupby('credit_history')['savings_status'].count()
cred = cred.reset_index()
ref[1].pie(cred['savings_status'], labels = cred['credit_history'])
ref[1].set_title("b. High Savings Breakdown by Credit History")

job = data.groupby('job')['savings_status'].count()
job = job.reset_index()
ref[2].pie(job['savings_status'], labels = job['job'])
ref[2].set_title("c. High Savings Breakdown by Job")

plt.show()
```
